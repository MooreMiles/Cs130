```python
ifndef MAIN_HPP
#define MAIN_HPP

#include <iostream>
#include <vector>

using namespace std;

void print(vector<int>);
void merge(vector<int> &num, int start, int mid, int end);
void merge_sort_recur(vector<int> &num, int first, int last);
void merge_sort_iter(vector<int> &num);

void print(vector<int> numbers)
{
    for (auto v : numbers)
        cout << v << "\t";
    cout << endl;
}


void merge(vector<int> &num, int start, int mid, int end)
{
 int lf = start;
    int rf = mid + 1;

    vector<int> merged;

    while (lf <= mid && rf <= end)
    {
        if (num[lf] < num[rf])
            merged.push_back(num[lf++]);
        else
            merged.push_back(num[rf++]);
    }

    while (lf <= mid)
        merged.push_back(num[lf++]);

    while (rf <= end)
        merged.push_back(num[rf++]);

    for (int i = 0; i < merged.size(); i++)
        num[start + i] = merged[i];
}

void merge_sort_recur(vector<int> &num, int first, int last)
{

    if (first >= last)
        return;

    int mid = (first + last) / 2;

    merge_sort_recur(num, first, mid);
    merge_sort_recur(num, mid + 1, last);
    merge(num, first, mid, last);
}

void merge_sort_iter(vector<int> &num)
{
    int n = (int)num.size();
    for (int size = 1; size < n; size *= 2)
    {
        for (int start = 0; start < n - size; start += 2 * size)
        {
            int mid = start + size - 1;
            int end = min(start + 2 * size - 1, n - 1);
            merge(num, start, mid, end);
        }
    }
}

#endif
```

```python
%%writefile main.hpp
#ifndef MAIN_HPP
#define MAIN_HPP

#include <iostream>
#include <vector>
#include <algorithm>

using namespace std;

void print(vector<int>);
void merge(vector<int> &num, int start, int mid, int end);
void merge_sort_recur(vector<int> &num, int first, int last);
void merge_sort_iter(vector<int> &num);

void print(vector<int> numbers)
{
    for (auto v : numbers)
        cout << v << "\t";
    cout << endl;
}

void merge(vector<int> &num, int start, int mid, int end)
{
    int lf = start;
    int rf = mid + 1;

    vector<int> merged;

    while (lf <= mid && rf <= end)
    {
        if (num[lf] < num[rf])
            merged.push_back(num[lf++]);
        else
            merged.push_back(num[rf++]);
    }

    while (lf <= mid)
        merged.push_back(num[lf++]);

    while (rf <= end)
        merged.push_back(num[rf++]);

    for (int i = 0; i < merged.size(); i++)
        num[start + i] = merged[i];
}

void merge_sort_recur(vector<int> &num, int first, int last)
{

    if (first >= last)
        return;

    int mid = (first + last) / 2;

    merge_sort_recur(num, first, mid);
    merge_sort_recur(num, mid + 1, last);
    merge(num, first, mid, last);
}

void merge_sort_iter(vector<int> &num)
{
    int n = (int)num.size();
    for (int size = 1; size < n; size *= 2)
    {
        for (int start = 0; start < n - size; start += 2 * size)
        {
            int mid = start + size - 1;
            int end = min(start + 2 * size - 1, n - 1);
            merge(num, start, mid, end);
        }
    }
}

#endif
```

```python
%%writefile main.cpp
#include "main.hpp"

// main function to test recursive and iterative merge sort
int main()
{
    vector<int> numbers_recur = {12, 11, 13, 5, 6, 7, 1, 9, 2, 8, 4, 10};
    cout << "Original array (Recursive Merge Sort): ";
    print(numbers_recur);
    merge_sort_recur(numbers_recur, 0, numbers_recur.size() - 1);
    cout << "Sorted array (Recursive Merge Sort): ";
    print(numbers_recur);

    cout << endl;

    vector<int> numbers_iter = {12, 11, 13, 5, 6, 7, 1, 9, 2, 8, 4, 10};
    cout << "Original array (Iterative Merge Sort): ";
    print(numbers_iter);
    merge_sort_iter(numbers_iter);
    cout << "Sorted array (Iterative Merge Sort): ";
    print(numbers_iter);

    return 0;
}
```
